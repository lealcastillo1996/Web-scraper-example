```python
#Importing used libraries
import pandas as pd
import ssl


#Importing ssl context
ssl.create_default_https_context = ssl._create_unverified_context

#pd html reader (inserting url to scrap)
scraper = pd.read_html("https://www.aa.com/es-mx/vuelos?reportedLocation=HomePageB1&reportedTitle=&_locale=es_MX#")


#Initiating a cycle to print all tables scraped
for idx,table in enumerate(scraper):
    
    
    print("----------------------------------")
    print("Table number:",idx)
    print(table)
    print(type(scraper))

#Select an specific table
scraper[0]
```
